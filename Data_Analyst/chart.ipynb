```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
plt.style.use('seaborn-whitegrid')
plt.plot()
```

```python
x=[1,2,3,4]
y=[10,8,6,9]
plt.plot(x,y,color='red')
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='blue',linestyle='dashed',label='sin(x)')
plt.plot(x,np.cos(x),color='red',label='cos(x)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.axis("equal")
plt.title('Sin(X) and Cos(X)')
plt.legend()
```

```python
x=[1,2,3,4]
y=[11,22,33,44]

fig,ax = plt.subplots(figsize=(5,5))

ax.plot(x,y)
ax.set(title="A simple plot",
              xlabel="x-axis",
              ylabel="y-axis"
              )
```

```python
rng = np.random.RandomState(100)

x = rng.randn(100)
y = rng.randn(100)
color = rng.randn(100)
size = 1000*rng.randn(100)

fig, ax = plt.subplots()
img = ax.scatter(x,y,s=size,c=color,cmap='viridis',alpha=0.3)

fig.colorbar(img)
```

```python
soft_drink_prices = {"Coke": 10,
                     "Pessi": 12,
                     "Sprite": 8}


fig, ax = plt.subplots()

ax.barh(list(soft_drink_prices.keys()),list( soft_drink_prices.values()))
ax.set(title="Bach Hoa Xah", ylabel="Price($)")
```
